```python
from pyspark.sql.functions import col

tables_default = spark.sql("SHOW TABLES IN default")
tables_default.filter(col("tableName").contains("silver")).show(truncate=False)
tables_default.filter(col("tableName").contains("gold")).show(truncate=False)
```
